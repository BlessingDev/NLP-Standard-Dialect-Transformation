```python
from metric import compute_bleu_score

def baseline_bleu_whole(region, token):
    file_path = "D:/박준호/2024년/standard-dialect/model_storage/dia_to_sta/{0}/{1}/prediction.json".format(region, token)
    
    prediction_list = []
    with open(file_path, mode="r+", encoding="utf-16") as fp:
        prediction_list = json.loads(fp.read())
    
    df = pd.DataFrame(prediction_list)
    
    bleu = compute_bleu_score(df["source"].to_list(), df["truth"].to_list())
    
    return bleu
    
```

```python
def baseline_bleu_dialect(region, token):
    file_path = "D:/박준호/2024년/standard-dialect/model_storage/dia_to_sta/{0}/{1}/prediction.json".format(region, token)
    
    prediction_list = []
    with open(file_path, mode="r+", encoding="utf-16") as fp:
        prediction_list = json.loads(fp.read())
    
    df = pd.DataFrame(prediction_list)
    df = df.loc[df["truth"] != df["source"]]
    
    bleu = compute_bleu_score(df["source"].to_list(), df["truth"].to_list())
    
    return bleu
```

영어 논문 Table token_result의 번역 없이 bleu 측정한 경우에 사용

```python
def prediction_bleu_dialect(region, token):
    file_path = "D:/박준호/2024년/standard-dialect/model_storage/dia_to_sta/{0}/{1}/prediction.json".format(region, token)
    
    prediction_list = []
    with open(file_path, mode="r+", encoding="utf-16") as fp:
        prediction_list = json.loads(fp.read())
    
    df = pd.DataFrame(prediction_list)
    df = df.loc[df["truth"] != df["source"]]
    
    bleu = compute_bleu_score(df["pred"].to_list(), df["source"].to_list())
    
    return bleu
```

```python
regions = ["gyeongsang", "jeonla", "jeju", "gangwon", "chungcheong"]
tokens = ["SentencePiece", "bpe", "형태소", "jamo"]

for r in regions:
    print(r)
    for t in tokens:
        print(t)
        bleu = prediction_bleu_dialect(r, t)
        print(bleu)
```
